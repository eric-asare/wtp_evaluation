```python
answer_file = open("/content/opus_bn.txt", 'r')
system_file = open("/content/temp_wtp.txt", 'r')
answer_key = []
system_key = []
pos = 0
for sentence in answer_file:
  pos += len(sentence) - 1
  if pos not in answer_key:
    answer_key += [pos]
pos = 0
for sentence in system_file:
  pos += len(sentence) - 1
  if pos not in system_key:
    system_key += [pos]
```

```python
def clean(text):
    clean_data = ""
    for character in text:
        if character == "\n":
            clean_data += " "
        else:
            clean_data += character
    return clean_data
```

```python
import xml.etree.ElementTree as ET

file_path = '/content/wtp/a-s.xml'  # Replace this with the actual file path

try:
    # Parse the XML file
    tree = ET.parse(file_path)
    root = tree.getroot()

    # Define the namespace used in the XML file
    namespace = {'graf': 'http://www.xces.org/ns/GrAF/1.0/'}

    # Find the region tags with anchors attribute
    regions = root.findall(".//graf:region[@anchors]", namespaces=namespace)

    if not regions:
        print("No regions found with anchors attribute.")
    else:
        for region in regions:
            anchors_value = region.get("anchors")
            print(f"Region found with anchors attribute: {anchors_value}")

except FileNotFoundError:
    print(f"File '{file_path}' not found.")
except ET.ParseError as e:
    print(f"Error parsing XML: {e}")
```
