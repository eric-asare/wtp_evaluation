```python
from wtpsplit import WtP
wtp = WtP("wtp-bert-mini")
wtp.half().to("cuda")
```

```python
import os
from collections import defaultdict
```

```python
os.mkdir("/content/WtP_System/")
os.mkdir("/content/WtP_System/data/")
os.mkdir("/content/WtP_System/data/written/")
```

```python
mount = "/content/drive/MyDrive"
directory = mount + "/MASC-3.0.0/data/written/"
```

```python
folders = os.listdir(directory)
```

```python
for folder in folders:
  if os.listdir(directory+folder):
    foundOne = False
    for item in os.listdir(directory+folder):
      if "." not in item:
        foundOne = True
        folders.append(f"{folder}/{item}")
    if foundOne:
      folders.remove(folder)

files = defaultdict(dict)
```

```python
for folder in folders:
  if folder.startswith('.'):
    continue
  files[folder][".xml"] = []
  files[folder][".txt"] = []
  for file in os.listdir(directory+folder):
    if "-s.xml" in file:
      files[folder][".xml"].append(file)
    elif ".txt" in file:
      files[folder][".txt"].append(file)
```

```python
print(files)
```

```python
for folder in folders:
  if "/" in folder:
    folder = folder.split("/")
    if not os.path.isdir("/content/WtP_System/data/written/" + folder[0]):
      os.mkdir("/content/WtP_System/data/written/" + folder[0])
    os.mkdir("/content/WtP_System/data/written/" + folder[0] + "/" + folder[1])
```

```python
for folder in folders:
  if "/" not in folder:
    os.mkdir("/content/WtP_System/data/written/" + folder)
```

```python
for genre in files:
  text_files = files[genre]['.txt']
  for text_file in text_files:
    input_file = open(directory + genre + "/" + text_file, 'r')
    input = input_file.read()
    output = wtp.split(input)
    pos = 0
    ends = []
    for sentences in output:
      ends += [(pos+len(sentences.rstrip()))]
      pos += len(sentences)
    output_file_name = "/content/WtP_System/data/written/" + genre + "/" + text_file.replace(".txt","-system.txt")
    output = open(output_file_name, 'w')
    for end in ends:
      print(end, file = output)
```
